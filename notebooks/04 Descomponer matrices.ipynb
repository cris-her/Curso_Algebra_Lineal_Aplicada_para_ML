```
import numpy as np
```

```
A = np.array([[3,2], [4,1]])
print(A)
```

```
autovalores, autovectores = np.linalg.eig(A)
print(autovalores)
print(autovectores)
```

```
A_calc = autovectores.dot(np.diag(autovalores)).dot(np.linalg.inv(autovectores))
print(A_calc)
```

Ejemplo 2 matriz simetrica real

```
B =  np.array([[3,2], [2,3]])
print(B)
```

```
print(B == B.T)
```

```
autovalores, autovectores = np.linalg.eig(B)
print(autovalores)
print(autovectores)
```

```
# recordamos que para una matriz simetrica su inversa es igual a su traspuesta
# esto es computacionalmente mas economico que una inversa
B_calc = autovectores.dot(np.diag(autovalores)).dot(autovectores.T)
print(B_calc)
```
