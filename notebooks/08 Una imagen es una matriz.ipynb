```
plt.style.use('classic')

imagen = Image.open('./imagen_ejemplo_frida_bredesen.jpg')

plt.imshow(imagen)
```

```
# Conversion imagen escala grises
imagen_gr = imagen.convert('LA')
print(imagen_gr)
```

```
imagen_mat = np.array(list(imagen_gr.getdata(band=0)), float)
print(imagen_mat)
```

```
# damos la forma que necesitamos a la matriz (para nosotros es un vector))
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
```

```
print(imagen_mat.shape)
```

```
plt.imshow(imagen_mat, cmap = 'gray')
```

```
imagen_mat_2 =  imagen_mat / 10
print(imagen_mat_2)
```

```
plt.imshow(imagen_mat_2)
plt.show()
```

Al haber dividido entre 10 estamos obteniendo una escala  de colores totalmente diferente de colores. Es necesario consevear esta escala o con tener la relacion de los valores basta.

Leamos la imagen usando un cmap de escala de grises

```
plt.imshow(imagen_mat_2, cmap = 'gray')
```
