```
import numpy as np

np.set_printoptions(suppress=True)
```

```
A = np.array([[2,3],[5,7],[11,13]])
print(A)
```

La peseudo inversa de Moore Penrose cuando verifica 4 condiciones existe y es unica

```
U,D,V = np.linalg.svd(A)

print('U')
print(U)

print("D")
print(D)

print("V")
print(V)
```

```
D_pse = np.zeros((A.shape[0],A.shape[1])).T
D_pse 
```

Al calcular la pseudo inversa hacemos el inverso de cada elemento (1/d11, 1/d12 ...)

```
print("Valores a reemplazar en D_pse")
print(D_pse[:D.shape[0], :D.shape[0]])
```

```
print("Valores que pondremos en D_pse")
print(np.linalg.inv(np.diag(D)))
```

```
print("D_pse")
D_pse[:D.shape[0], :D.shape[0]] = np.linalg.inv(np.diag(D))

print(D_pse)
```

Construirmos la pseudoinversa A

```
A_pse = V.T.dot(D_pse).dot(U.T)
print(A_pse)
```

Para validar esto calculemos la psudoinversa con numpy

```
A_pse_calc = np.linalg.pinv(A)
print(A_pse_calc)
```

```
print(A_pse == A_pse_calc)
```

```
print(A_pse.dot(A))
```

```
print(A_pse_calc.dot(A))
```

Observamos que tenemos un -0, esto porque tenemos minimizados los valores cercanos a cero

```
np.set_printoptions(suppress=False)
```

```
print(A_pse.dot(A))
```

- Observamos que la matriz no es exactamente la inversa, estamos teniendo algo muy cercano, ya que minimiza la norma 2 de estas distancias (de estos errores que estamos cometiendo).

- Otra propiedad es que estas matrices no son conmutativas

```
print(A.dot(A_pse))
```

Y cuando intentamos invertir las matrices obtenemos algo muy alejado de la identidad, tambien hay otras pseudoinversas, la mas usada es la de SVD la de Moore Penrose pero podemos escribir otra y obtener un valor igual o muy aproximado

```
A_pse_2 = np.linalg.inv(A.T.dot(A)).dot(A.T)
print(A_pse_2)
```
