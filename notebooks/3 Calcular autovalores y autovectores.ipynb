```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
X = np.array([[3,2], [4,1]])
print(X)
```

Calculo de autovalores y autovectores

```
print(np.linalg.eig(X))
```

```
autovalores, autovectores = np.linalg.eig(X)
print(autovalores)
```

```
print(autovectores[:,0])
```

```
print(autovectores[:,1])
```
