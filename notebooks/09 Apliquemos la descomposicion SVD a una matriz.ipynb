```
plt.style.use('classic')

imagen = Image.open('./imagen_ejemplo_frida_bredesen.jpg')

plt.imshow(imagen)
```

```
# Conversion imagen escala grises
imagen_gr = imagen.convert('LA')

# Convertirla en una matriz
imagen_mat = np.array(list(imagen_gr.getdata(band=0)), float)

# Convertirmos el vector imagen_mat en una matriz
# tomando la referencia de la imagen original size=3693x3456
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
```

Aplicamos descomposicion SVD

```
U, D, V = np.linalg.svd(imagen_mat)
```

```
print(imagen_mat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

Recordemos que D tiene los valores singulares, estos estan ordenados de manera tal
que el primer valor es el que contiene mayor informacion, es el que contiene mas  varianza de los datos

```
imagen_recons = np.matrix(U[:, :1]) * np.diag(D[:1]) * np.matrix(V[:1,:])
plt.imshow(imagen_recons, cmap='gray')
```

Tenemos la matriz representada como una imagen, pero cuantos valores singulares tendriamos que utilizar para que la imagen se vuelva reconocible.

### Buscando la cantidad de valores singulares que nos sirvan
