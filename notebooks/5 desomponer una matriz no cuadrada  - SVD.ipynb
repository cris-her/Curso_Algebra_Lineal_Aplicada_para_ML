```
import numpy as np
```

```
A = np.array([[1,2,3],[3,4,5]])
print(A)
```

```
U, D, V = np.linalg.svd(A)
```

```
print(U)
```

```
print(D)
```

```
print(np.diag(D))
```

```
print(V)
```

**Contenido adicional solucion A = UDV**

UDV deben ser de dimensiones compatibles para realizar producto interno, asi que agregamos una columna con ceros al final

```
D_aumentada = np.hstack((np.diag(D),np.zeros((2,1))))
print(D_aumentada)
```

```
A_calc = U.dot(D_aumentada).dot(V)
print(A_calc)
```
